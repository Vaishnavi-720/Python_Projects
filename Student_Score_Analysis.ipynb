```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#information about the data
df.info()
```

# Data Transformation

### Dropping Unnamed column

```python
#dropping unnamed column
df = df.drop("Unnamed: 0", axis = 1)
```

```python
#printing top 5
print(df.head())
```

### changing values in wklyStudyHours column

```python
df["WklyStudyHours"] = df["WklyStudyHours"].str.replace("< 5", "< 5 Hrs")
print(df.head())
```

```python
df["WklyStudyHours"] = df["WklyStudyHours"].str.replace("5 - 10", "5 - 10 Hrs")
print(df.head())
```

```python
df.head()
```

### Detecting Outliers

```python
#Detecting outliers in maths scores using Box plot

sns.boxplot(data = df, x = 'MathScore')
plt.show()
```

```python
#Detecting outliers in reading scores using Box plot

sns.boxplot(data = df, x = 'ReadingScore')
plt.show()
```

### From above box plot we can see that ReadingScore contains Outliers

```python
#Detecting outliers in writing scores using Box plot

sns.boxplot(data = df, x = 'WritingScore')
plt.show()
```

### From above box plot we can see that WritingScore contains Outliers
    Average WritingScore lies between 65 to 69

# Creating charts

### Gender Distribution

```python
#counting total numbers of students based on gender using countplot
plt.figure(figsize = (4,5))
ax = sns.countplot(data = df, x = "Gender")
plt.title("Gender Distribution")

#showing datalabels
ax.bar_label(ax.containers[0])
plt.show()
```

### Insights from above chart:

####  from above chart we have analysed that:
     1. We can see that there is marginal difference in gender distribution
     2. Number of female is more than number of male
     3. Count of female student is 15424 whereas count of male student is 15217

```python
#Does parent's education imapct onstudent's score

gb = df.groupby("ParentEduc").agg({"MathScore": "mean", "ReadingScore": "mean", "WritingScore": "mean"})
print(gb)
```

```python
plt.figure(figsize = (4,4))
sns.heatmap(gb, annot = True)
plt.title("Impact of Parent's Eduaction")
plt.show()
```

### from above chart we have analysed that:
    
    1. Parent's education do imapact on student's score
    2. Student's scored higher marks whose parents had done master's degree

```python
#Does Parent's maritial status have impact on student's score

gbm = df.groupby("ParentMaritalStatus").agg({"MathScore": "mean", "ReadingScore": "mean", "WritingScore": "mean"})
print(gbm)
```

```python
plt.figure(figsize = (4,4))
sns.heatmap(gbm, annot = True)
plt.title("Impact of Parent's Marital Status")
plt.show()
```

### from above chart we have analysed that:
    
    1. parent's marital status is not much affecring student's score i.e. the impact is negligible
    2. Average of students score lies between 66 to 70 marks

## Distribution of Ethnic Group

```python
GroupA = df.loc[(df["EthnicGroup"] == "group A")].count()
GroupB = df.loc[(df["EthnicGroup"] == "group B")].count()
GroupC = df.loc[(df["EthnicGroup"] == "group C")].count()
GroupD = df.loc[(df["EthnicGroup"] == "group D")].count()
GroupE = df.loc[(df["EthnicGroup"] == "group E")].count()

l = ["group A", "group B", "group C", "group D", "group E"]
mlist = [GroupA["EthnicGroup"], GroupB["EthnicGroup"], GroupC["EthnicGroup"], GroupD["EthnicGroup"], GroupE["EthnicGroup"]]

plt.pie(mlist, labels = l, autopct = "%1.2f%%")
plt.title("Distribution of Ethnic Groups")
plt.show()
```

### From above chart we can say that :
    1. Group C has maximum distribution with 31.98 %
    2. Whereas Group A has minimum distribution with 7.70%
